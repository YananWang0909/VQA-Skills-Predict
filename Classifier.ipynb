```python
#In the prototype, we focus on "text recognition" only 
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model.logistic import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
kfold=KFold(n_splits=10)
```

```python
# reading in the CSV files as Pandas dataframes
df_train_ori = pd.read_csv('vizwiz_skill_typ_train.csv', skipinitialspace = True, engine='python')
df_val_ori = pd.read_csv('vizwiz_skill_typ_val.csv', skipinitialspace = True, engine='python')
# The structure of VizWiz file
df_train_ori.head(50)

```

```python
df_train_ori.ix[1,0]
a = df_train_ori[(df_train_ori.CNT==1)].index.tolist()
print(a)
```

```python
# reading in the CSV files as Pandas dataframes
api_feature_train = pd.read_csv('api_features_train.csv', skipinitialspace = True, engine='python')
api_feature_val = pd.read_csv('api_features_val.csv', skipinitialspace = True, engine='python')
nn_feature_train = pd.read_csv('nn_features_train.csv', skipinitialspace = True, engine='python')
nn_feature_val = pd.read_csv('nn_features_val.csv', skipinitialspace = True, engine='python')
# The structure of VizWiz file
api_feature_train.head(3)
```

```python
df_train = pd.merge(df_train_ori, api_feature_train, how='left', on='qid')
df_train = pd.merge(df_train, nn_feature_train, how='left', on='qid')
df_train.head(3)
df_val = pd.merge(df_val_ori, api_feature_val, how='left', on='qid')
df_val = pd.merge(df_val, nn_feature_val, how='left', on='qid')
df_val.head(3)
```

```python
#merge text
df_train["allText1"] = df_train["QSN"].map(str) + df_train["tags"].map(str)
df_train["allText2"] = df_train["QSN"].map(str) + df_train["vgg_tags alexnet_tags"].map(str)
df_train["allText3"] = df_train["QSN"].map(str) + df_train["descriptions"].map(str)
df_train.head(3)
df_val["allText1"] = df_val["QSN"].map(str) + df_val["tags"].map(str)
df_val["allText2"] = df_val["QSN"].map(str) + df_val["vgg_tags alexnet_tags"].map(str)
df_val["allText3"] = df_val["QSN"].map(str) + df_val["descriptions"].map(str)
df_val.head(3)
```

```python
#create X and Ys
ques_train=df_train["QSN"].values
txt_train=df_train["TXT"].values
obj_train=df_train["OBJ"].values
col_train=df_train["COL"].values
cnt_train=df_train["CNT"].values

ques_val=df_val["QSN"].values
txt_val=df_val["TXT"].values
obj_val=df_val["OBJ"].values
col_val=df_val["COL"].values
cnt_val=df_val["CNT"].values
```

```python
#bag of words
from sklearn.feature_extraction.text import CountVectorizer
#do not move stopwords as they are possibly important
#count=CountVectorizer(stop_words='english')
count=CountVectorizer()
bag_train=count.fit_transform(ques_train)#create vocabulary
bag_val=count.transform(ques_val)
print(count.vocabulary_)
```

```python
quesBOWFeatures_train=bag_train.toarray()
quesBOWFeatures_val=bag_val.toarray()
```

```python
#TF-IDF
from sklearn.feature_extraction.text import TfidfTransformer
tfidf=TfidfTransformer(use_idf=True,
                      norm= 'l2',
                      smooth_idf=True)
np.set_printoptions(precision=2)
tfidfResultTrain=tfidf.fit_transform(bag_train)
tfidfResultVal=tfidf.transform(bag_val)
```

```python
quesTFIDFFeatures_train=tfidfResultTrain.toarray()
quesTFIDFFeatures_val=tfidfResultVal.toarray()
print(quesTFIDFFeatures_train.shape)
print(quesTFIDFFeatures_val.shape)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from itertools import cycle

from sklearn.metrics import roc_curve, auc
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import label_binarize
from sklearn.multiclass import OneVsRestClassifier
from scipy import interp
```

```python
#Try several models to predict
#X is question features, Y is whether it need text recognition
from sklearn.metrics import accuracy_score, roc_auc_score, roc_curve
NaiveBayes=GaussianNB()
KNN=KNeighborsClassifier()
Tree=DecisionTreeClassifier()
LR=LogisticRegression()


NaiveBayes.fit(quesTFIDFFeatures_train,col_train)
testScore=NaiveBayes.score(quesTFIDFFeatures_val,col_val)
print("Guassian Naive Bayes")
print("Test Score:{:.2f}\n".format(testScore))
preds = NaiveBayes.predict(quesTFIDFFeatures_val)
print(("Accuracy = {0} \t AUC = {1}".format(accuracy_score(col_val, preds), 
       roc_auc_score(col_val, preds))))


KNN.fit(quesTFIDFFeatures_train,col_train)
testScore=KNN.score(quesTFIDFFeatures_val,col_val)
print("KNN")
print("Test Score:{:.2f}\n".format(testScore))
preds = KNN.predict(quesTFIDFFeatures_val)
print(("Accuracy = {0} \t AUC = {1}".format(accuracy_score(col_val, preds), 
       roc_auc_score(col_val, preds))))



Tree.fit(quesTFIDFFeatures_train,col_train)
testScore=Tree.score(quesTFIDFFeatures_val,col_val)
print("Tree")
print("Test Score:{:.2f}".format(testScore))
preds = Tree.predict(quesTFIDFFeatures_val)
print(("Accuracy = {0} \t AUC = {1}".format(accuracy_score(col_val, preds), 
       roc_auc_score(col_val, preds))))

LR.fit(quesTFIDFFeatures_train,col_train)
testScore=LR.score(quesTFIDFFeatures_val,col_val)
print("Logistic Regression")
print("Test Score:{:.2f}".format(testScore))
preds = LR.predict(quesTFIDFFeatures_val)
print(("Accuracy = {0} \t AUC = {1}".format(accuracy_score(col_val, preds), 
       roc_auc_score(col_val, preds))))
```
